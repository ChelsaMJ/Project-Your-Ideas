```python
# calculating weight:
def calc_bmi(weight, height):
    # BMI = weight/(height**2)*703
    bmi = weight / (height ** 2) * 703
    return bmi

#calculating BMI:
def level_bmi(bmi):
    if bmi < 16:
        return "Severely Underweight"
    elif 16 <= bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi < 25:
        return "Healthy"
    elif 25 <= bmi < 30:
        return "Overweight"
    else:
        return "Obese"

#Calorie data
def calc_cal_consumption(food_items):
    calorie_data = {
        "Milk": 100,
        "Egg": 155,
        "Rice": 130,
        "Lentils": 113,
        "Vegetable": 85,
        "Meat": 143
    }

#Adding up calories consumed:
    total_calories = 0
    for item, quantity in food_items.items():
        if item in calorie_data:
            total_calories += (calorie_data[item] * quantity)

    return total_calories

#BODY
def main():
    name = str(input("Enter your name: "))
    age = int(input("Enter your age: "))
    weight = float(input("Enter your weight in pounds: "))
    height = float(input("Enter your height in inches: "))

    bmi = calc_bmi(weight, height)
    level = level_bmi(bmi)

    print("BMI of", name, "is", bmi, "and he/she is", level)

#Calories required for range 0-8
    recommended_calories = 0
    if 0 <= age <= 2:
        recommended_calories = 800
    elif 2 < age <= 4:
        recommended_calories = 1400
    elif 4 < age <= 8:
        recommended_calories = 1800

#Taking input from user about items consumed:
    food_items = {
        "Milk": float(input("Enter quantity of Milk (in grams): ")),
        "Egg": float(input("Enter quantity of Egg (in grams): ")),
        "Rice": float(input("Enter quantity of Rice (in grams): ")),
        "Lentils": float(input("Enter quantity of Lentils (in grams): ")),
        "Vegetable": float(input("Enter quantity of Vegetable (in grams): ")),
        "Meat": float(input("Enter quantity of Meat (in grams): "))
    }

    consumed_calories = calc_cal_consumption(food_items)

    print("Recommended daily calorie requirement:", recommended_calories, "calories")
    print("Total consumed calories:", consumed_calories, "calories")

#checking whether the child is undernourished:
    if consumed_calories < recommended_calories:
        print("The child is undernourished.")
    else:
        print("The child is not undernourished.")

main()
```
