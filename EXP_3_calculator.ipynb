```python
#functions for different operations
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y != 0:
        return x / y
    else:
        return "Cannot divide by zero"

def hcf(x, y):
    while y:
        x, y = y, x % y
    return x

def lcm(x, y):
    lcm_value = (x * y) // hcf(x, y)
    return lcm_value

def power(x, y):
    result = x ** y
    return result

def log(x, base):
    import math
    result = math.log(x, base)
    return result


# Input from the user
print("Options:")
print("1. Add")
print("2. Subtract")
print("3. Multiply")
print("4. Divide")
print("5. HCF")
print("6. LCM")
print("7. Power")
print("8. Logarithm")

#choosing an operation
choice = input("Enter choice (1/2/3/4/5/6/7/8): ")

#Performing user's choice:
if choice in ('1', '2', '3', '4', '7', '8'):
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))

if choice == '1':
    print("Result:", add(num1, num2))
elif choice == '2':
    print("Result:", subtract(num1, num2))
elif choice == '3':
    print("Result:", multiply(num1, num2))
elif choice == '4':
    if num2 == 0:
        print("Cannot divide by zero")
    else:
        print("Result:", divide(num1, num2))
elif choice == '5':
    print("Result:", hcf(int(num1), int(num2)))
elif choice == '6':
    print("Result:", lcm(int(num1), int(num2)))
elif choice == '7':
    print("Result:", power(num1, num2))
elif choice == '8':
    base = float(input("Enter the base for logarithm: "))
    if num1 <= 0 or base <= 0:
        print("Invalid input for logarithm")
    else:
        print("Result:", log(num1, base))
else:
    print("Invalid input")



```
